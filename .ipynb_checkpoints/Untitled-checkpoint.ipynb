## Modeling

```python
data.shape
```

```python
# Split the data into X and y
X = data.drop("quality", axis = 1)
y = data["quality"]

# Split X and y into Train and test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20)

# Check the shape of the splits
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Store model scores for comparaing
scores_df = {"name": [],
             "train_score": [],
             "test_score":[]}

# Define function for evaluating base scores
def base_score(model_info):
    model,name = model_info
    model.fit(X_train, y_train)
    scores_df["name"].append(name)
    scores_df["train_score"].append(model.score(X_train, y_train))
    scores_df["test_score"].append(model.score(X_test, y_test))
    
# Models to train
models = [(KNeighborsClassifier(), "knn"),
          (SVC(), "svc"),
          (RandomForestClassifier(), "rfs"),
          (BernoulliNB(),"bnb"),
          (GaussianNB(), "gnb"),
          (LinearDiscriminantAnalysis(), "lda"),
          (LogisticRegression(), "lg")]
```

```python
# Train models
for i in models:
    base_score(i)
```

```python
scores_df = pd.DataFrame(scores_df)
scores_df
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax = sns.barplot(x = "name", y = "train_score", data = scores_df, label = "Model")
plt.title("Model's train score - accuracy", pad = 30 , fontdict = {"fontsize":20,
                                                       "fontweight":20})
ax.set(xlabel = "Model", ylabel = "Score")
ax.set_ylim(top = 1);
```

```python
fig, ax = plt.subplots(figsize = (10,5))
ax = sns.barplot(x = "name", y = "test_score", data = scores_df, label = "Model")
plt.title("Model's test score - accuracy", pad = 30 , fontdict = {"fontsize":20,
                                                       "fontweight":20})
ax.set(xlabel = "Model", ylabel = "Score")
ax.set_ylim(top = 1);
```

### Randomized search

```python
model_rfc = RandomForestClassifier()
model_rfc.get_params()
```

```python
model_lr = LogisticRegression()
model_lr.get_params()
```

```python
model_lda = LinearDiscriminantAnalysis()
model_lda.get_params()
```

```python
model_knn = KNeighborsClassifier()
model_knn.get_params()
```

```python
grid = {
    model_rfc:
    {
        "n_estimators":range(100,1500,200),
        "max_features":["auto", "sqrt"],
        "max_depth":range(5,30,5),
        "min_samples_split":[2, 5, 10],
        "min_samples_leaf":[1, 2, 4],
        "bootstrap":[True,False]
    },
    model_lr:
    {
        "solver":["newton-cg", "lbfgs", "liblinear"],
        "penalty": ["l1", "l2"],
        "C": np.arange(0.1, 1, 0.1) 
    },
    model_lda:
    {
        "solver": ["lsqr","eigen"],
        "shrinkage": [0.1,1,0.1]
    },
    model_knn:
    {
        "leaf_size": range(1,50),
        "n_neighbors": range(1,30),
        "p":[1,2]
    }
}
```

```python
best_estimators = []
```

```python
for model in grid.keys():
    clf = GridSearchCV(estimator = model,
                       param_grid = grid[model],
                       cv = 5,
                       verbose = 2,
                       n_jobs = -1)
    train = clf.fit(X_train, y_train)
    best_estimators.append(train.best_estimator_)
```

```python
def scoring(estimator):
    estimator.fit(X_train, y_train)
    print(estimator.score(X_test, y_test))
    
for estimator in best_estimators:
    print(estimator)
    scoring(estimator)
```

```python
model_rfc = RandomForestClassifier(max_depth=25, n_estimators=500)
model_lr = LogisticRegression(C=0.30000000000000004)
model_lda = LinearDiscriminantAnalysis(shrinkage=0.1, solver='lsqr')
model_knn = KNeighborsClassifier(leaf_size=1, n_neighbors=17, p=1)
```

```python
model_rfc.fit(X_train, y_train)
model_lr.fit(X_train, y_train)
model_lda.fit(X_train, y_train)
model_knn.fit(X_train, y_train)
```

```python
print(f"Random Forest Classifier accuracy score is: {model_rfc.score(X_test, y_test)*100:.2f}")
print(f"Logistic regression accuracy score is: {model_lr.score(X_test, y_test)*100:.2f}")
print(f"Linear Discriminant Analysis accuracy score is: {model_lda.score(X_test, y_test)*100:.2f}")
print(f"KNN is: {model_knn.score(X_test, y_test)*100:.2f}")
```

```python
# Helper function for plotting feature importance
def plot_features(columns, importances, n=20):
    df = (pd.DataFrame({"features":columns,
                       "feature_importances":importances})
         .sort_values("feature_importances", ascending = False)
         .reset_index(drop = True))
    
    fig,ax = plt.subplots(figsize = (10,10))
    ax.barh(df["features"][:n],df["feature_importances"][:20])
    ax.set_ylabel("Features")
    ax.set_xlabel("Feature importance")
    ax.invert_yaxis()
```

```python
plot_features(X.columns, model_rfc.feature_importances_ )
```
